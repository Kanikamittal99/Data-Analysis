```python
import pandas as pd
```

```python
data = pd.read_csv('adult.data.csv')
data.head()
```

**1. How many men and women (*sex* feature) are represented in this dataset?** 

```python
#(data['sex'] == 'Male').value_counts()
data.groupby('sex')['sex'].count()
```

**2. What is the average age (*age* feature) of women?**

```python
# You code here
data.groupby('sex')['age'].mean()
```

**3. What is the percentage of German citizens (*native-country* feature)?**

```python
# count_rows = data[data['native-country'] == 'Germany'].shape[0]

German_citizens = len(data[data['native-country'] == 'Germany'])
total_pop = len(data['native-country'])
perc_german = (German_citizens/total_pop)*100
print(perc_german)
```

**4-5. What are the mean and standard deviation of age for those who earn more than 50K per year (*salary* feature) and those who earn less than 50K per year? **

```python
print(data[data['salary'] == '>50K']['age'].mean())
print(data[data['salary'] == '>50K']['age'].std())
```

```python
print(data[data['salary'] == '<=50K']['age'].mean())
print(data[data['salary'] == '<=50K']['age'].std())
```

**6. Is it true that people who earn more than 50K have at least high school education? (*education â€“ Bachelors, Prof-school, Assoc-acdm, Assoc-voc, Masters* or *Doctorate* feature)**

```python
true, false = 0,0
df  = data[data['salary'] == '>50K']
for x in df['education']:
    if ((x =='Bachelors') or (x == 'Assoc-acdm' ) or (x =='Assoc-voc') or (x == 'Masters') or (x == 'Doctorate')):
        true = true+1
    else:
        false = false+1
print("true: {} false: {}".format(true,false))
```

**7. Display age statistics for each race (*race* feature) and each gender (*sex* feature). Use *groupby()* and *describe()*. Find the maximum age of men of *Amer-Indian-Eskimo* race.**

```python
Max = data.loc[(data['race'] == 'Amer-Indian-Eskimo') & (data['sex'] == 'Male')]['age'].max() 
print("Maximum age of men of Amer-Indian_Eskimo race: ",Max)

data.groupby(['race','sex'])[['age']].describe()
```

**8. Among whom is the proportion of those who earn a lot (>50K) greater: married or single men (*marital-status* feature)? Consider as married those who have a *marital-status* starting with *Married* (Married-civ-spouse, Married-spouse-absent or Married-AF-spouse), the rest are considered bachelors.**

```python
df = data[data['salary'] == '>50K']
t = len(df.index)
m,s = 0,0
for x in df['marital-status']:
    if ((x == 'Married-civ-spouse') or (x == 'Married-spouse-absent') or (x == 'Married-AF-spouse')):
        m = m+1
    else:
        s = s+1

print("married: {}%  single: {}%".format(round((m/t)*100, 2), round((s/t)*100,2)))
```

**9. What is the maximum number of hours a person works per week (*hours-per-week* feature)? How many people work such a number of hours, and what is the percentage of those who earn a lot (>50K) among them?**

```python
a = data['hours-per-week'].max()
b = len(data[data['hours-per-week'] == 99])
num = len(data.loc[(data['hours-per-week'] == 99) & (data['salary'] == '>50K')])
c = (num/b)*100


print("Max. no. of hours person works per week: {}".format(a))
print("{} people work such a no. of hours.".format(b))
print("{}% of above people earn >50K".format(round(c,2)))
```
